```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Plotting Sigmoidal Activation function
h = np.linspace(-5,5,50)
x=tf.placeholder(tf.float32)
out = tf.sigmoid(x)
dout = tf.gradients(out,x)
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    y, dY = sess.run([out, dout], feed_dict={x:h})
    gradient=dY[0]
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Sigmoidal Activation Function and Its Derivative')
plot1 = plt.plot(h, y, label = 'Sigmoid Function')
plot2 = plt.plot(h, gradient, label = 'Sigmoid Derivative')
plt.legend(loc='upper left')
```

```python
# Threshold Activation function
def threshold (x):
    cond = tf.less(x, tf.zeros(tf.shape(x), dtype = x.dtype))
    out = tf.where(cond, tf.zeros(tf.shape(x)), tf.ones(tf.shape(x)))
    return out
```

```python
# Plotting Threshold Activation function
h = np.linspace(-5, 5, 50)
x=tf.placeholder(tf.float32)
out = threshold(x)
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out, feed_dict={x:h})
    gradient=dY[0]
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Threshold Activation Function')
plot1 = plt.plot(h, y, label = 'Threshold Function')
plt.legend(loc='upper left')
```
