```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(5,10, 50)
y = np.linspace(5,10, 50)
x, y = np.meshgrid(x, y)
```

```python
z = x * np.sin(4*x) + 1.1* np.sin (2*y)
```

```python
fig = plt.figure(figsize=(30,30))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("X", fontsize=20)
ax.set_ylabel("Y", fontsize=20)
ax.set_zlabel("Surface", fontsize=20)
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1,
                cmap='RdYlBu', edgecolor='none')#ax.plot_surface(x, y, z,cmap=cm.coolwarm, linewidth=0.1, antialiased=True)
#fig.colorbar(surf, shrink=0.5, aspect=5)
#ax.view_init(35, 60)
```

```python
plt.contourf(x, y, z,  30, cmap='RdYlBu')
plt.xlabel("X", fontsize=20)
plt.ylabel("Y", fontsize=20)
plt.colorbar();
```
